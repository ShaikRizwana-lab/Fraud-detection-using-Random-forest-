# Understanding Dataset & Cleaning

```python
# Get unique values from multiple columns
unique_values_dict = {}
for column in df.columns:
    unique_values_dict[column] = df[column].unique()

# Print unique values for each column
for column, unique_values in unique_values_dict.items():
    print(f"Unique values in '{column}': {unique_values}")
```

```python
print(df.isnull().sum())
print('\n')
print("total rows in the df -")
print(len(df))
```

```python
#Handling the null values in the dataset

df['Deductible'] = df['Deductible'].fillna(df['Deductible'].mean())

# round the mean to a whole number instead of float to stick to the scale provided (1 to 4)
mean_driver_rating = df['DriverRating'].mean()
rounded_mean_driver_rating = round(mean_driver_rating)
df['DriverRating'] = df['DriverRating'].fillna(rounded_mean_driver_rating)

columns_to_check = ['Make', 'Sex', 'PoliceReportFiled','WitnessPresent','FraudFound_P'] #try mode
df.dropna(subset=columns_to_check, inplace=True)

df.isnull().sum()

```

```python
# Replace 'MALE' with 'male' in the 'Sex' column
df['Sex'] = df['Sex'].replace('MALE', 'Male')

# Replace 'MALE' with 'male' in the 'gender' column
df['Sex'] = df['Sex'].replace('FEMALE', 'Female')
```

```python
# Define a function to convert different date formats to ISO 8601 and standardise them 
def convert_to_iso(date):
    try:
        return pd.to_datetime(date).strftime('%Y-%m-%d')
    except ValueError:
        try:
            return pd.to_datetime(date, format='%d %B %Y', errors='raise').strftime('%Y-%m-%d')
        except ValueError:
            return pd.to_datetime(date, errors='coerce').strftime('%Y-%m-%d')

# Apply the function to the 'RegistrationDate' column
df['RegistrationDate'] = df['RegistrationDate'].apply(convert_to_iso)
```

```python
df['RegistrationDate_Year'] = pd.to_datetime(df['RegistrationDate']).dt.year
df['RegistrationDate_Month'] = pd.to_datetime(df['RegistrationDate']).dt.month
df['RegistrationDate_Day'] = pd.to_datetime(df['RegistrationDate']).dt.day

df
```

```python
# Locate the rows where "DriverRating" is equal to -2
rows_to_update = df['DriverRating'] == -2

# Update the values in the "DriverRating" column where condition is True
df.loc[rows_to_update, 'DriverRating'] = 2

# Print the DataFrame to verify the change
print(df)
```
